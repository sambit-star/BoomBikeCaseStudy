# STEP 2: Preparing the data for modelling
    - Encoding:
        -Converting binary variables to 1/0( this is already done in data set)
        -Other categorical variables to dummy ones

# Dropping the unimportant variables instant,dteday,casual,registered and workingday column

```python
boombikedf.drop('instant',axis=1, inplace=True)
boombikedf.drop('dteday',axis=1, inplace=True)
boombikedf.drop('casual',axis=1, inplace=True)
boombikedf.drop('registered',axis=1, inplace=True)
#boombikedf.drop('workingday',axis=1, inplace=True)
boombikedf.head()
```

```python
## yes/no variables fare already in boolean 1/0 format 
varlist = ['yr','holiday','workingday']

boombikedf[varlist].head()
```

# Create dummy variables to represent month, Season, Weekday, weathersit

```python
monthvar = ['January','February','March','April','May','June','July','August','September','October','November','December']
seasonvar = ['spring', 'summer', 'fall','winter']
weekdayvar = ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
#weathersit : 
#		- 1: Clear, Few clouds, Partly cloudy, Partly cloudy --> GOOD
#		- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist --> MISTY
#		- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds --> SNOWY
#		- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog --> BAD
weathervar = ['Good','Misty','Snowy','Bad']
```

```python
# Month starts from 1, Season starts from 1, Weekday starts from 0, weathersit starts from 1
boombikedf['weathercategory'] = boombikedf['weathersit'].apply(lambda x: weathervar[x-1])
boombikedf['monthcategory'] = boombikedf['mnth'].apply(lambda x: monthvar[x-1])
boombikedf['seasoncategory'] = boombikedf['season'].apply(lambda x: seasonvar[x-1])
boombikedf['weekdaycategory'] = boombikedf['weekday'].apply(lambda x: weekdayvar[x])
boombikedf.head()
```

```python
# Drop original variables
boombikedf.drop('weathersit',axis=1, inplace=True)
boombikedf.drop('mnth',axis=1, inplace=True)
boombikedf.drop('season',axis=1, inplace=True)
boombikedf.drop('weekday',axis=1, inplace=True)
boombikedf.head()
```

```python
#Create the dummy variables
monthdummy = pd.get_dummies(boombikedf['monthcategory'],drop_first=True).astype(int)
weatherdummy = pd.get_dummies(boombikedf['weathercategory'],drop_first=True).astype(int)
seasondummy = pd.get_dummies(boombikedf['seasoncategory'],drop_first=True).astype(int)
weekdaydummy = pd.get_dummies(boombikedf['weekdaycategory'],drop_first=True).astype(int)

boombikedf = pd.concat([boombikedf,monthdummy,weatherdummy,seasondummy,weekdaydummy],axis=1)

#Drop temporary variables
boombikedf.drop('monthcategory',axis=1, inplace=True)
boombikedf.drop('weathercategory',axis=1, inplace=True)
boombikedf.drop('seasoncategory',axis=1, inplace=True)
boombikedf.drop('weekdaycategory',axis=1, inplace=True)

boombikedf.head()
```

## STEP 3: Split dataset into train and test

```python

from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(boombikedf, train_size=0.7, random_state=100)
print(df_train)
print(df_test)
```

## STEP 4: Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
num_vars = ['temp','atemp','hum','windspeed','cnt']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])

df_train.head()
```

## Dividing into X and Y sets for the model building

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

## STEP 5: Building Model

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, n_features_to_select=15)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```

```python
X_train_rfe.columns
```

```python
X_train_new = X_train_rfe.drop(['const'], axis=1)
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## hum is insignificant so dropping (>5)

```python
X_train_new = X_train_rfe.drop(["hum"], axis = 1)
```

## Rebuilding model without hum

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_lm = sm.add_constant(X_train_new)
```

```python
lm = sm.OLS(y_train,X_train_lm).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```

```python
X_train_lm.columns
```

```python
X_train_new = X_train_lm.drop(["const"], axis = 1)
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

# Removing workingday due high VIF

```python
X_train_new = X_train_lm.drop(["workingday"], axis = 1)

# Adding a constant variable 
import statsmodels.api as sm  
X_train_lm = sm.add_constant(X_train_new)

lm = sm.OLS(y_train,X_train_lm).fit()   # Running the linear model

#Let's see the summary of our linear model
print(lm.summary())

X_train_lm.columns

X_train_new = X_train_lm.drop(["const"], axis = 1)

# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Removing Saturday due to high p value(>.05)

```python
X_train_new = X_train_lm.drop(["Saturday"], axis = 1)

# Adding a constant variable 
import statsmodels.api as sm  
X_train_lm = sm.add_constant(X_train_new)

lm = sm.OLS(y_train,X_train_lm).fit()   # Running the linear model

#Let's see the summary of our linear model
print(lm.summary())

X_train_lm.columns

X_train_new = X_train_lm.drop(["const"], axis = 1)

# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Keeping back the temp though its >=5 mark for VIF since in pairplot we see that temp has a good correlation on the 'cnt'

```python
y_train_cnt = lm.predict(X_train_lm)
```

# Get the coefficients and feature names

```python
coefficients = lm.params
sorted_coefficients = coefficients.abs().sort_values(ascending=False)
print(sorted_coefficients)
```

```python
# Use our model to make predictions.

# Creating X_test_new dataframe by dropping variables from X_test
X_test_new = X_test[X_train_new.columns]

# Adding a constant variable 
X_test_new = sm.add_constant(X_test_new)
```

```python
# Making predictions
y_pred = lm.predict(X_test_new)
```
